```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
```

Normalization is done so that all feature are consistent to each other, it scale the data between 0 and 1. 
formula is z = (X-Xmin)/(Xmax-Xmin)

```python
# Get dataset
df = pd.read_csv("data/california_housing_train.csv", sep=",")
```

```python
df.head()
```

```python
# Normalize total_bedrooms column
x_array = np.array(df['total_bedrooms'])
```

```python
normalized_X = preprocessing.normalize([x_array])
```

```python
print(x_array)
print(normalized_X)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
sns.distplot(df['total_bedrooms'], bins =100)
plt.show()
```

```python
sns.distplot(df['median_house_value'], bins =100)
plt.show()
```

```python
sns.distplot(normalized_X, bins =100)
plt.show()
```

```python
x__array = np.array(df['median_house_value'])
normalized__X = preprocessing.normalize([x__array])
sns.distplot(normalized__X, bins =100)
plt.show()
```

In standarization we keep the value center around 0 then between 0 and 1.
Formula z = (Xi -Mean)/sigma


```python
# Create the Scaler object
scaler = preprocessing.StandardScaler()
```

```python
scaled_df = scaler.fit_transform(df)
```

```python
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
```

```python
scaled_df
```

```python
sns.distplot(scaled_df['total_bedrooms'], bins =100)
plt.show()
```

```python
sns.distplot(scaled_df['median_house_value'], bins =100)
plt.show()
```
